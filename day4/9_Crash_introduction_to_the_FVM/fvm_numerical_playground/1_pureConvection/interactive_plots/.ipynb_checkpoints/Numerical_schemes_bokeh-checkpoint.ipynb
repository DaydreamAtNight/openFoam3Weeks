```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

#from bokeh.models import HoverTool, BoxSelectTool

output_notebook()
#output_file("experiment1.html",mode='inline')



TOOLS = 'pan,box_zoom,crosshair,resize,reset,hover,wheel_zoom,save,help'
#TOOLS = [BoxSelectTool(), HoverTool(),box_zoom]



data=np.loadtxt('sol_exact/exact.txt', skiprows=0)
data0=np.loadtxt('sol0/line_centreProfile_T.xy.0', skiprows=0)
data1=np.loadtxt('sol0/line_centreProfile_T.xy.1', skiprows=0)
data2=np.loadtxt('sol0/line_centreProfile_T.xy.2', skiprows=0)
data3=np.loadtxt('sol0/line_centreProfile_T.xy.3', skiprows=0)
data4=np.loadtxt('sol0/line_centreProfile_T.xy.4', skiprows=0)
data5=np.loadtxt('sol0/line_centreProfile_T.xy.5', skiprows=0)
data6=np.loadtxt('sol0/line_centreProfile_T.xy.6', skiprows=0)


p = figure(title='Solution at time t = 0.5 s', x_axis_label='Length', y_axis_label='Scalar T',
          plot_width=800, plot_height=800,webgl=True,tools=TOOLS)

p.line(data[:,0], data[:,1], legend="Exact solution", line_width=2)

p.line(data0[:,0], data0[:,1], legend="Upwind", line_width=2,line_color="red")
p.line(data1[:,0], data1[:,1], legend="Linear-Upwind", line_width=2,line_color="orange")
p.line(data2[:,0], data2[:,1], legend="Linear", line_width=2,line_color="green")
p.line(data3[:,0], data3[:,1], legend="QUICK", line_width=2,line_color="cyan")
p.line(data4[:,0], data4[:,1], legend="TVD-vanLeer", line_width=2,line_color="black")
p.line(data5[:,0], data5[:,1], legend="TVD-minmod", line_width=2,line_color="magenta")
p.line(data6[:,0], data6[:,1], legend="TVD-SuperBee", line_width=2,line_color="brown")


show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
from bokeh.palettes import Spectral11

output_notebook()
#output_file("experiment2.html",mode='inline')

data=np.loadtxt('sol_exact/exact.txt', skiprows=0)
data0=np.loadtxt('sol1/line_centreProfile_T.xy.0', skiprows=0)
data1=np.loadtxt('sol1/line_centreProfile_T.xy.1', skiprows=0)
data2=np.loadtxt('sol1/line_centreProfile_T.xy.2', skiprows=0)
data3=np.loadtxt('sol1/line_centreProfile_T.xy.3', skiprows=0)
data4=np.loadtxt('sol1/line_centreProfile_T.xy.4', skiprows=0)
data5=np.loadtxt('sol1/line_centreProfile_T.xy.5', skiprows=0)

mypalette=Spectral11[0:6]


p = figure(title='Solution at time t = 0.5 s', x_axis_label='Length', y_axis_label='Scalar T',
          plot_width=800, plot_height=800)


#linear upwind

p.line(data[:,0], data[:,1], legend="Exact solution", 
       line_width=2)

p.line(data0[:,0], data0[:,1], legend="Gauss linear", 
       line_width=2,line_color="red")
p.line(data1[:,0], data1[:,1], legend="cellMDLimited Gauss linear 1.0", 
       line_width=2,line_color="orange")
p.line(data2[:,0], data2[:,1], legend="cellMDLimited Gauss linear 0.5", 
       line_width=2,line_color="green",line_dash=[8, 8])
p.line(data3[:,0], data3[:,1], legend="cellMDLimited Gauss linear 0", 
       line_width=2,line_color="cyan",line_dash=[6, 6])
p.line(data4[:,0], data4[:,1], legend="cellLimited leastSquares 1.0", 
       line_width=2,line_color="black",line_dash=[4, 4])
p.line(data5[:,0], data5[:,1], legend="cellLimited leastSquares 0.333", 
       line_width=2,line_color="magenta",line_dash=[4, 4])


show(p)
```
